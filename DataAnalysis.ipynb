Filter Methods Used:
     Correlation Coefficient 
 

```python
#FILTER METHOD
#1. Correlation Coefficient
#using features that are highly directly correlated
plt.figure(figsize=(14,7))
coef = df.corr()['Result'].sort_values()[::-1][1:9]
plt.bar(coef.index, coef.values)
```

```python
from sklearn.model_selection import train_test_split
X = df[['SSLfinal_State', 'URL_of_Anchor', 'Prefix_Suffix', 'web_traffic', 'having_Sub_Domain',
        'Request_URL','Links_in_tags','SFH']]
y = df['Result']
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=15)
rfc.fit(X_train, y_train)
print('Random Forest Classifier')
print('Training Score:',rfc.score(X_train, y_train))
print('Testing Score:',rfc.score(X_test, y_test))
```

```python
predicted = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
para = {'C' : [0.1,1,10,100],'gamma' : [0.1,0.01,0.001,0.0001]}
gcv = GridSearchCV(SVC(),param_grid = para, verbose =0)

#C and gamma are basically responsible bias and variance of the model
#A large value of C will give low bias and high variance
#A large value of gamma will give high bias and low variance 
#in order to determine the best possible combination of C and gamma we use GridSearchCV
```

```python
gcv.fit(X_train,y_train)
gcv.best_params_
gcv.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = 'liblinear')
lr.fit(X_train, y_train)
print('Logistic Regression')
print('Training Score:',lr.score(X_train, y_train))
print('Testing Score:',lr.score(X_test, y_test))
```

```python
lr_predicted = lr.predict(X_test)
print(classification_report(y_test,lr_predicted))
```
